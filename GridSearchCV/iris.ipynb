```python
from sklearn.datasets import load_iris
df=load_iris()
dir(df)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.data,df.target,test_size=0.2)
len(x_train)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
import pandas as pd
clf=GridSearchCV(SVC(gamma='auto'),{'C':[1,10,20],'kernel':['rbf','linear']},cv=5,return_train_score=False)
clf.fit(df.data,df.target)
res=pd.DataFrame(clf.cv_results_)
res
```

```python
res[['params','mean_test_score']]
```

```python
clf.best_params_
```

```python
from sklearn.model_selection import RandomizedSearchCV
rs=RandomizedSearchCV(SVC(gamma='auto'),{
    'C':[1,10,20],'kernel':['rbf','linear']
},cv=5,return_train_score=False,n_iter=3)
rs.fit(df.data,df.target)
res2=pd.DataFrame(rs.cv_results_)
res2
```

```python
res2[['params','mean_test_score']]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
model={
    'svm':{
        'model':SVC(gamma='auto'),
        'params':{
            'C':[1,10,20],
            'kernel':['rbf','linear']
        }
    },
    'random_forest':{
        'model':RandomForestClassifier(),
        'params':{
            'n_estimators':[1,5,10]
        }
    },
    'Logistic_Regression':{
        'model':LogisticRegression(solver='liblinear',multi_class='auto'),
        'params':{
            'C':[1,5,10]
        }
    }
}
```

```python
scores=[]
for mn,mp in model.items():
    clf2=GridSearchCV(mp['model'],mp['params'],cv=5,return_train_score=False)
    clf2.fit(df.data,df.target)
    scores.append({
        'model':mn,
        'best_score':clf2.best_score_,
        'best_params':clf2.best_params_
    })
scores
```
