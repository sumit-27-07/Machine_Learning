```python
import pandas as pd
df=pd.read_csv(r"C:\Users\sumit\Desktop\Code\python\Machine Learning\Principle component\heart.csv")
df.head()
```

```python
df.shape
```

```python
df['zscr_bp']=(df.RestingBP-df.RestingBP.mean())/df.RestingBP.std()
df1=df[(df.zscr_bp<3) & (df.zscr_bp>-3)]
df1.head()
```

```python
df1['zscr_ch']=(df1.Cholesterol-df1.Cholesterol.mean())/df1.Cholesterol.std()
df2=df1[(df1.zscr_ch<3) & (df1.zscr_ch>-3)]
df2.head()
```

```python
df2.shape
```

```python
df2['zscr_hr']=(df2.MaxHR-df2.MaxHR.mean())/df2.MaxHR.std()
df3=df2[(df2.zscr_hr<3) & (df2.zscr_hr>-3)]
df3.head()
```

```python
df3['zscr_op']=(df3.Oldpeak-df3.Oldpeak.mean())/df3.Oldpeak.std()
df4=df3[(df3.zscr_op<3) & (df3.zscr_op>-3)]
df4.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df4.ChestPainType=le.fit_transform(df4.ChestPainType)
df4
```

```python
df4.Sex=le.fit_transform(df4.Sex)
df4.RestingECG=le.fit_transform(df4.RestingECG)
df4.ExerciseAngina=le.fit_transform(df4.ExerciseAngina)
df4.ST_Slope=le.fit_transform(df4.ST_Slope)
df4
```

```python
y=df4.HeartDisease
y
```

```python
df4
```

```python
x=df4.drop(['HeartDisease','zscr_bp','zscr_ch','zscr_hr','zscr_op'],axis='columns')
x
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_scale=sc.fit_transform(x)
x_scale
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scale,y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
import numpy as np
```

```python
scr_lg=cross_val_score(LogisticRegression(),x_train,y_train,cv=5)
print("LogisticRegression",scr_lg)
print(np.average(scr_lg))
scr_svc=cross_val_score(SVC(),x_train,y_train,cv=5)
print("SVM",scr_svc)
print(np.average(scr_svc))
scr_rf=cross_val_score(RandomForestClassifier(),x_train,y_train,cv=5)
print("Random Forest",scr_rf)
print(np.average(scr_rf))
```
